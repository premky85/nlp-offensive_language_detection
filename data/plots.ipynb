```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

class_map = {0: "non-offensive", 1: "offensive", 2:"hate", 3: "both"}
filenames = ['fox_news_comments', 'gab', 'reddit', 'slo_test_data', 'slo_twitter_binary', 'twitter', 'white_supremacist_forum']
titles = ['Fox news comments', 'Gab comments', 'Reddit comments', 'Slo data', 'Slo twitter', 'Eng twitter', 'White supremacist forum']
```

```python
for filename,title in zip(filenames, titles):
    df = pd.read_csv('preprocessed_data/' + filename + '_preprocessed.csv')
    distr = dict(Counter([class_map[c] for c in df['class']]))

    plt.figure(figsize=(6,5))
    plt.pie(distr.values(), labels=distr.keys(), shadow=True, autopct='%1.1f%%')
    plt.title(title + ' class distribution')
    # plt.show()
    plt.savefig('plots/' + filename + '.png')
```
